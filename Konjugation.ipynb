# Konjugation
Almanca'da düzenli fiiller, çekimleri belirli bir kalıp veya düzen içinde alır. Bu düzen, fiilin kökünün (Stamm) sonuna eklenen belirli eklerle belirtilir. İşte Almanca'da düzenli fiillerin çekimlenmesi hakkında bazı gramatik kurallar:

1. Fiil Kökü (Stamm):
   - Düzenli bir Almanca fiilin temel formu, fiil kökü veya "Stamm" olarak adlandırılır. Fiil kökü, fiilin anlamını taşıyan kısmıdır ve ekler bu köke eklenir.

2. Şahıs Zamirleri (Personel Pronomen):
   - Fiil çekimleri, cümlenin öznesine (şahıs zamiri) bağlı olarak değişir. Almanca'da şahıs zamirleri şunlardır: "ich" (ben), "du" (sen), "er/sie/es" (o), "wir" (biz), "ihr" (siz), "sie/Sie" (onlar/siz).

3. Çekim Ekleri (Endungen):
   - Düzenli fiiller, şahıs zamirlerine bağlı olarak farklı çekim ekleri alırlar. Çekim ekleri, fiil köküne eklenir ve fiilin zaman, kişi ve sayıya göre nasıl çekildiğini belirtir.
   - İşte düzenli bir Almanca fiilin şahıs zamirlerine göre çekim ekleri örnekleri:
     - "ich" için: "-e" eki eklenir (sprechen - ich spreche)
     - "du" için: "-st" eki eklenir (sprechen - du sprichst)
     - "er/sie/es" için: "-t" eki eklenir (sprechen - er/sie/es spricht)
     - "wir" için: "-en" eki eklenir (sprechen - wir sprechen)
     - "ihr" için: "-t" eki eklenir (sprechen - ihr sprecht)
     - "sie/Sie" için: "-en" eki eklenir (sprechen - sie/Sie sprechen)

4. Zaman (Tense):
   - Almanca fiiller, genellikle geçmiş zaman (Präteritum) ve şimdiki zaman (Präsens) olmak üzere farklı zamanlarda çekimlenir. Bazı fiiller ayrıca gelecek zaman (Futur) ve diğer zamanlar için de çekim alabilirler.

5. Dil Bilgisi Kuralları:
   - Düzenli fiiller, Almanca dil bilgisi kurallarına uygun olarak çekimlenir. Bu kurallar, fiilin sonuna eklenen çekim eklerini belirler ve fiilin doğru bir şekilde çekilmesini sağlar.

Örnek:
- "sprechen" fiili (konuşmak), düzenli bir Almanca fiilidir. Şimdiki zaman (Präsens) çekimleri aşağıdaki gibidir:
  - ich spreche (ben konuşuyorum)
  - du sprichst (sen konuşuyorsun)
  - er/sie/es spricht (o konuşuyor)
  - wir sprechen (biz konuşuyoruz)
  - ihr sprecht (siz konuşuyorsunuz)
  - sie/Sie sprechen (onlar/siz konuşuyor)
